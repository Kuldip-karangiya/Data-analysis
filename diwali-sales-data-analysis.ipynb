```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Diwali-Sales-Data.csv',encoding='unicode_escape')
df.shape
```

```python
df.head(10)
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'],axis = 1, inplace = True)
```

```python
pd.isnull(df).sum()
```

```python
df.shape
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df['Amount'] = df['Amount'].astype('int')
```

```python
df.describe()
```

```python
df[['Age','Orders','Amount']].describe()
```

```python
#Exploratory data analysis
```

```python
df.columns
```

```python
ax = sns.countplot(x = 'Gender',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Gender'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
```

```python
sales_gen = df.groupby(['Gender'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
sns.barplot(x = 'Gender',y = 'Amount',data = sales_gen)
```

```python
ax = sns.countplot(x = 'Age Group', hue = 'Gender',data = df)
for bars in ax.containers:
    ax.bar_label(bars)
    
```

```python
sales_age = df.groupby(['Age Group'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)
sns.barplot(x = 'Age Group', y = 'Amount', data = sales_age)
```

```python
sales_state = df.groupby(['State'],as_index = False)['Orders'].sum().sort_values(by = 'Orders',ascending = False)

sns.set(rc = {'figure.figsize' : (25,6)})
sns.barplot(x = 'State', y = 'Orders',data = sales_state)
```

```python
sales_satat_amount = df.groupby(['State'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)

sns.barplot(x = 'State', y = 'Amount',data= sales_satat_amount)
```

```python
df.columns
```

```python
ax = sns.countplot(x = 'Marital_Status', hue = 'Gender', data = df)

sns.set(rc = {'figure.figsize' : (7,5)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax = sns.countplot(x = 'Occupation',data = df)

sns.set(rc = {'figure.figsize' : (25,6)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occupation = df.groupby(['Occupation'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)

sns.barplot(x = 'Occupation', y = 'Amount',data= sales_occupation)
```

```python
ax = sns.countplot(x = 'Product_Category',data = df)

sns.set(rc = {'figure.figsize' : (25,6)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_product_category = df.groupby(['Product_Category'],as_index = False)['Amount'].sum().sort_values(by = 'Amount',ascending = False)

sns.barplot(x = 'Product_Category', y = 'Amount',data= sales_product_category)
```

```python
most_saling_product = df.groupby(['Product_ID'],as_index = False)['Orders'].sum().sort_values(by = 'Orders',ascending = False).head(10)

sns.set(rc = {'figure.figsize' : (20,5)})
sns.barplot(x = 'Product_ID', y = 'Orders',data= most_saling_product)
```
